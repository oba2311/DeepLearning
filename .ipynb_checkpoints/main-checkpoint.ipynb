```python
import numpy as np
import pandas as pd
# I'll use pandas to manipulate data, but not to write the algo...

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"


# loading the reacy, binarized data:
Xtraining = pd.read_csv('/Users/a8407352/Desktop/deepLearn/datasets/titanic/trainReady.csv', header = 0,dtype={'1': np.float64})
Xtraining = Xtraining.drop(Xtraining.columns[0],axis=1)
```

```python
age_avg=Xtraining['0'].mean()
age_std=Xtraining['0'].std()
age_null_count =Xtraining['0'].isnull().sum()
age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
Xtraining['0'][np.isnan(Xtraining['0'])] = age_null_random_list
Xtraining['0'] = Xtraining['0'].astype(int)

Xtraining['CategoricalAge'] = pd.cut(Xtraining['0'], 5)

print (Xtraining[['CategoricalAge', '16']].groupby(['CategoricalAge'], as_index=False).mean())

Xtraining.loc[ Xtraining['0'] <= 16, '0']                          = 0
Xtraining.loc[(Xtraining['0'] > 16) & (Xtraining['0'] <= 32), '0'] = 1
Xtraining.loc[(Xtraining['0'] > 32) & (Xtraining['0'] <= 48), '0'] = 2
Xtraining.loc[(Xtraining['0'] > 48) & (Xtraining['0'] <= 64), '0'] = 3
Xtraining.loc[ Xtraining['0'] > 64, '0']                           = 4
```

```python
# Xtraining=Xtraining.drop(Xtraining['CategoricalAge'], inplace=True)
Xtraining = Xtraining.iloc[:,0:25]
Xtraining.head(10)
```

```python
yTrain = pd.read_csv('/Users/a8407352/Desktop/deepLearn/datasets/titanic/trainLabelsReady.csv')
yTrain = yTrain.drop(yTrain.columns[0],axis=1)
yTrain.head()
```

```python
def sigmoid(x, deriv=False):
    if deriv:
        return x*(1-x)
    return 1/(1+np.exp(-x))

    
weights1 = np.random.random((Xtraining.shape[1],1)) #bias can be added here, like -1 for example..
weights2 = np.random.random((1,1))  #L1.shape[0]
```

```python
errors = []
index = []
for batch in range(10000):
    L0 = Xtraining
    L1 = sigmoid(Xtraining.dot(weights1)) #891X1
    L2 = sigmoid(L1.dot(weights2)) 
#     more layers here...possibly.
#         error: 
    er = yTrain.values-L2        # 891X1
    index.append(batch)
    errors.append(np.mean(np.abs(er))) #since we take the mean of a batch of errors, we would have less variation
    #within the results, or smoothen results.
    if batch%100==0:
        print ('error:' +str(np.mean(np.abs(er))))

        
        
#         learning / backprop:
#     propogate:
#      regular vector multuplication to find the delta:
    delta1 = er*sigmoid(L2, deriv=True) # 2X1
    er2 =  delta1.dot(weights2.T)  # 2X2
    delta2 = er2 * sigmoid(L1, deriv=True) # 2X2
#     update weights with gradient descent:
    weights2 += L2.T.dot(delta1)
    weights1 += L1.T.dot(delta2)
    
```

```python
weights1 = np.random.random((Xtraining.shape[1],1)) #bias can be added here, like -1 for example..
weights2 = np.random.random((1,1))  #L1.shape[0]
```

```python
def Relu(x, deriv=False):
    if deriv:
        x[x<=0] = 0
        x[x>0] = 1
        return x #notice that the derivative is not defined for x=0, but we'll use an approximation here.
    else:
        return np.maximum(0,x)
```

```python
errors = []
index = []
for batchNum in range(100):
    L0 = Xtraining
    dotProduct = Xtraining.dot(weights1)
    L1 = Relu(dotProduct) #891X1
    L2 = Relu(L1.dot(weights2))

# obviously, we see by now that ReLu is not adequate for the task, 
# i.e. for binary classification we need to squeeze  the outpute into a 0 to 1 range,
# or in other words, probability.
```

```python
# Let's change the loss function - perhaps a log loss would be better.
def cross_entropy(predictions, targets):
    N = predictions.shape[0]
    ce = -np.sum(targets*np.log(predictions))/N
    return ce
```

```python
errors = []
index = []
for batch in range(10000):
    L0 = Xtraining
    L1 = sigmoid(Xtraining.dot(weights1)) #891X1
    L2 = sigmoid(L1.dot(weights2)) 
#     more layers here...possibly.
#         error: 
    er = cross_entropy(L2,yTrain.values)        # 891X1
    index.append(batch)
    errors.append(np.mean(np.abs(er)))
    if batch%100==0:
        print ('error:' +str(np.mean(np.abs(er))))

        
        
#         learning / backprop:
#     propogate:
#      regular vector multuplication to find the delta:
    delta1 = er*sigmoid(L2, deriv=True) # 2X1
    er2 =  delta1.dot(weights2.T)  # 2X2
    delta2 = er2 * sigmoid(L1, deriv=True) # 2X2
#     update weights with gradient descent:
    weights2 += L2.T.dot(delta1)
    weights1 += L1.T.dot(delta2)
    
```

```python
import matplotlib.pyplot as plt

def draw_neural_net(ax, left, right, bottom, top, layer_sizes):
    '''
    Draw a neural network cartoon using matplotilb.
    
    :usage:
        >>> fig = plt.figure(figsize=(12, 12))
        >>> draw_neural_net(fig.gca(), .1, .9, .1, .9, [4, 7, 2])
    
    :parameters:
        - ax : matplotlib.axes.AxesSubplot
            The axes on which to plot the cartoon (get e.g. by plt.gca())
        - left : float
            The center of the leftmost node(s) will be placed here
        - right : float
            The center of the rightmost node(s) will be placed here
        - bottom : float
            The center of the bottommost node(s) will be placed here
        - top : float
            The center of the topmost node(s) will be placed here
        - layer_sizes : list of int
            List of layer sizes, including input and output dimensionality
    '''
    n_layers = len(layer_sizes)
    v_spacing = (top - bottom)/float(max(layer_sizes))
    h_spacing = (right - left)/float(len(layer_sizes) - 1)
    # Nodes
    for n, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
        for m in range(layer_size):
            circle = plt.Circle((n*h_spacing + left, layer_top - m*v_spacing), v_spacing/4.,
                                color='w', ec='k', zorder=4)
            ax.add_artist(circle)
    # Edges
    for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing*(layer_size_a - 1)/2. + (top + bottom)/2.
        layer_top_b = v_spacing*(layer_size_b - 1)/2. + (top + bottom)/2.
        for m in range(layer_size_a):
            for o in range(layer_size_b):
                line = plt.Line2D([n*h_spacing + left, (n + 1)*h_spacing + left],
                                  [layer_top_a - m*v_spacing, layer_top_b - o*v_spacing], c='k')
                ax.add_artist(line)
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.gca()
ax.axis('off')
draw_neural_net(ax, .1, .9, .1, .9, [6, 6])
fig.savefig('nn.png')
```
